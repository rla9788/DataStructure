```python
큐 = [0,0,0,0,0,0]

앞쪽 = -1
뒤쪽 = -1

while True:
    print('큐 프로그램.. (1)삽입 (2)삭제 (3)출력 (4)종료')
    선택 = int(input(''))
    
    if 선택==1 :
        if 앞쪽 == 5:  # 큐가 가득 찼을 때
            print("큐가 가득 찼습니다. 추가가 불가능합니다.")
        else:
            앞쪽+=1
            큐[앞쪽] = int(input('값을 입력하세요'))
            
    elif 선택==2:
        if 앞쪽 == 뒤쪽:  # 큐가 비어있을 때
            print("큐가 비어있습니다. 삭제할 수 없습니다.")
        else:
            뒤쪽+=1
            print(f'{큐[뒤쪽]} 값을 삭제합니다')
            
    elif 선택==3:
        for i in range(뒤쪽+1, 앞쪽+1):
            print(큐[i], end=' ')
        print()
        
    else:
        break

    # 앞쪽과 뒤쪽이 같아졌을 때 큐 초기화
    if 앞쪽 == 뒤쪽:
        큐 = [0,0,0,0,0,0]
        앞쪽 = -1
        뒤쪽 = -1
        print("큐가 초기화되었습니다.")
```
